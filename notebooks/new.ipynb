```python
import numpy as np
from numpy import dot
from numpy.linalg import eigh, eig, norm, inv
import pandas as pd
import plotly.express as px
from plotly import graph_objects as go
# from scipy.optimize import curve_fit
import rampy

import importlib

import PCA
importlib.reload(PCA)
```

```python
roi = np.array([[1300,1350]])
# map_raw = PCA.Map('../data/mapa.txt', begin=1300, end=1385)
map = PCA.Map('../data/mapa.txt', begin=1300, end=1385, roi=roi, baseline='arPLS')
```

```python
from scipy.optimize import curve_fit

# Define the Lorentzian function
def lorentz(X, scale, mean, sigma):
    return scale / (np.pi * sigma * (1 + ((X - mean) / sigma) ** 2))

# Fit Lorentzian curve to each spectrum
params = []
for i, spectrum in enumerate(map.data_matrix):
    print(i)
    x_data = map.raman_shift
    y_data = spectrum
    popt = [0., 1368., 10.]
    try:
        popt, _ = curve_fit(lorentz, x_data, y_data, p0=[1, 1368, 10], bounds=([0, 1365, 0], [np.inf, 1371, 10]))
    except:
        pass
    params.append(popt)

params = np.array(params)

# Show map of fitted parameters
fig_amp = px.imshow(np.reshape(params[:, 0], map.shape), title='Amplitude')
fig_cen = px.imshow(np.reshape(params[:, 1], map.shape), title='Center')
fig_wid = px.imshow(np.reshape(params[:, 2], map.shape), title='Width')

fig_amp.show()
fig_cen.show()
fig_wid.show()
```
